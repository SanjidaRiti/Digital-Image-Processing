```python
import cv2
print("Original Image")
image=cv2.imread("flower4.jpg")
from matplotlib import pyplot as plt
plt.imshow(image)
```

```python
print("Gray scale")
image1=cv2.imread("flower4.jpg",0)
plt.imshow(image1)
```

```python
from numpy import asarray
data=asarray(image)
print(data)
```

```python
import imutils
print("Rotated Image")
print("90 angle rotation")
r1=imutils.rotate(image,90)
plt.imshow(r1)
```

```python
print("Original Image")
plt.imshow(image)
print("RGB Image")
R,G,B=cv2.split(image)
print("R")
plt.imshow(R)
print("G")
plt.imshow(G)
print("B")
plt.imshow(B)
```

```python
C=1-R
M=1-G
Y=1-B
print("C complacement")
plt.imshow(C)
print("M complacement")
plt.imshow(M)
print("Y complacement")
plt.imshow(Y)
```
